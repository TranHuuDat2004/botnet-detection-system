```python
import pandas as pd
import json

# Kaggle tự động mount dataset vào đường dẫn /kaggle/input/
# Thay 'botnet-cicids2017-cleaned' bằng tên dataset bạn đã đặt
file_path = '/kaggle/input/cicids2017-cleaned-data/cicids2017_cleaned.csv'

print(f"Đang đọc file dữ liệu từ: {file_path}")

# Đọc file CSV
try:
    df = pd.read_csv(file_path)
    print("Đọc file thành công!")

    # --- BẮT ĐẦU TÍNH TOÁN ---

    # 1. Lấy tổng số dòng trong toàn bộ dữ liệu
    total_rows = len(df)

    # 2. Sử dụng value_counts() - đây chính là "hàm collection" mạnh mẽ nhất để đếm
    # số lần xuất hiện của mỗi giá trị trong cột 'Attack Type'
    attack_type_counts = df['Attack Type'].value_counts()

    # 3. Lấy ra số lượng lưu lượng bình thường ('Normal Traffic')
    # Dùng .get(key, 0) để tránh lỗi nếu không có 'Normal Traffic'
    normal_count = attack_type_counts.get('Normal Traffic', 0)

    # 4. Tổng số lần tấn công = Tổng số dòng - số dòng bình thường
    attack_count = total_rows - normal_count

    # 5. Tạo một dictionary chi tiết các loại tấn công
    # Chúng ta sẽ loại bỏ 'Normal Traffic' ra khỏi danh sách thống kê
    attack_breakdown = attack_type_counts.drop('Normal Traffic', errors='ignore').to_dict()

    # --- TẠO DICTIONARY KẾT QUẢ CUỐI CÙNG ---
    final_stats = {
        "total_rows": total_rows,
        "normal_count": int(normal_count),
        "attack_count": int(attack_count),
        "attack_breakdown": attack_breakdown
    }

    # --- IN KẾT QUẢ RA DƯỚI DẠNG JSON ---
    # Sử dụng indent=4 để JSON được định dạng đẹp, dễ đọc và copy
    print("\n--- KẾT QUẢ DƯỚI DẠNG JSON ---")
    print(json.dumps(final_stats, indent=4))

except FileNotFoundError:
    print(f"LỖI: Không tìm thấy file tại đường dẫn '{file_path}'. Vui lòng kiểm tra lại tên dataset.")
except Exception as e:
    print(f"Đã xảy ra lỗi: {e}")
```
